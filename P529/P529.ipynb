```python
#Mit einem Plattenkondensator gemessener Ionisationsstrom für verschiedene 
#Kondensatorspannungen und bei 3 verschiedenen Röhrenspannungen (mit Kupfer Anode)

import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as optimize


                 
# Deklariere eine neue Funktion
def f(x, a, b):
    return a * x + b

def chi2(x, y, s, f, a, b):
    chi = (y - f(x, a, b)) / s
    return np.sum(chi**2)

# Ihre Messdaten kommen hier hin
data=np.loadtxt("ku15.txt",skiprows=1)
data2=np.loadtxt("ku27.5.txt",skiprows=1)
data3=np.loadtxt("ku30.txt",skiprows=1)


xdata1=data[:,0]
ydata1=data[:,2]
sigma1 = data[:,5] 
xerr=data[:,3] 

xdata2=data2[:,0]
ydata2=data2[:,2]
sigma2 = data2[:,5] 
xerr2=data2[:,3]

xdata3=data3[:,0]
ydata3=data3[:,2]
sigma3 = data3[:,5]
xerr3=data3[:,3]


# Startwerte für die Parameterbestimmung
x0 = np.array([1.0, 0.0])

fitParams1, fitCovariances1 = optimize.curve_fit(f, xdata1, ydata1, x0, sigma=sigma1, absolute_sigma=True)
fitParams2, fitCovariances2 = optimize.curve_fit(f, xdata2, ydata2, x0, sigma=sigma2, absolute_sigma=True)
fitParams3, fitCovariances3 = optimize.curve_fit(f, xdata3, ydata3, x0, sigma=sigma3, absolute_sigma=True)

slope1 = fitParams1[0]
intercept1 = fitParams1[1]

slope2 = fitParams2[0]
intercept2 = fitParams2[1]

slope3 = fitParams3[0]
intercept3 = fitParams3[1]

fitErrors1 = np.sqrt(np.diag(fitCovariances1))
fitErrors2 = np.sqrt(np.diag(fitCovariances2))
fitErrors3 = np.sqrt(np.diag(fitCovariances3))

thischi2_1 = chi2(xdata1, ydata1, sigma1, f, fitParams1[0], fitParams1[1])
thischi2_2 = chi2(xdata2, ydata2, sigma2, f, fitParams2[0], fitParams2[1])
thischi2_3 = chi2(xdata3, ydata3, sigma3, f, fitParams3[0], fitParams3[1])

# nun erschaffen wir einen plot
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111)

# Die Achsen brauchen auf jeden Fall eine Beschriftung
plt.title('Kupfer')
plt.ylabel('$I_C$[A]', fontsize=16)
plt.xlabel('$U_C$[V]', fontsize=16)
plt.xlim(0,280)
plt.ylim(0.00000000017, 0.00000003033)

# Plotte die Funktion
t = np.arange(25,100)

plt.errorbar(xdata1, ydata1,  xerr, fmt='mv', label='$U=15kV$')
plt.errorbar(xdata2, ydata2,  xerr2, fmt='rs', label='$U=27.5kV$')
plt.errorbar(xdata3, ydata3,  xerr3, fmt='co', label='$U=30kV$')
plt.grid(linestyle='-', color='0.9', linewidth=2)
plt.legend()

# Zeige den Plot
plt.show()

# Speichere den Plot
plotname = 'myplot.png'
plt.savefig(plotname, bbox_inches=0, dpi=600)
plt.close()

# Parameter ausgeben
print('Fit 1 - slope: {0:.3f} +- {1:.3f}, intercept: {2:.3f} +- {3:.3f}'.format(slope1, fitErrors1[0], intercept1, fitErrors1[1]))
print('Fit 2 - slope: {0:.3f} +- {1:.3f}, intercept: {2:.3f} +- {3:.3f}'.format(slope2, fitErrors2[0], intercept2, fitErrors2[1]))
print('Fit 3 - slope: {0:.3f} +- {1:.3f}, intercept: {2:.3f} +- {3:.3f}'.format(slope3, fitErrors3[0], intercept3, fitErrors3[1]))
print('Chi-squared values: Fit 1 - {0:.3f}, Fit 2 - {1:.3f}, Fit 3 - {2:.3f}'.format(thischi2_1, thischi2_2, thischi2_3))
```
